## Remover registros inconsistentes

Algunos registros en la base de datos no tienen mucho sentido, por ejemplo:

* Departamentos con un valor de alquiler igual a 0.

*  Departamentos con un valor de condominio igual a 0.

Estos registros son inconsistentes, por lo tanto, debemos eliminarlos de nuestra base de datos.

#### Consultamos los registros dónde el valor o el condominio es igual a 0.

```python
df.query('Valor == 0 | Condominio == 0')
```

Tenemos 500 registros donde se cumple la condición

#### Para ver los índices de dichos registros

```python
df.query('Valor == 0 | Condominio == 0').index
```

#### Vamos a traer todos los índices para eliminar fila por fila esos registros

```python
df_remover = df.query('Valor == 0 | Condominio == 0').index
```

#### Eliminamos los registros

**axis** es el eje, en este caso 0, porque eliminaremos filas del registro.

**inplace** es para que realice la modificación directamente sobre la variable **df**

```python
df.drop(df_remover, axis = 0, inplace = True)
```

```python
df.query('Valor == 0 | Condominio == 0').index
```

Vemos que la consulta nos devuelve una lista vacía, es decir que no hay registros cuyo Valor o Condominio tenga un valor de 0

```python
df.head()
```

#### Eliminamos columna Departemento por innecesaria

Como el único valor de Tipo que tenemos es Departamento, eliminaremos la columna

**axis** = 1 porque vamos a eliminar una columna

```python
df.drop('Tipo', axis = 1, inplace=True)
```

## Aplicar filtros

En este momento, el equipo de ML desea evaluar solo propiedades en escenarios específicos.

Con base en esto, necesitamos aplicar los filtros necesarios para obtener esta información específica.

### 1. Apartamentos que tienen 1 dormitorio y un alquiler menor a MXN 4200.

```python
seleccion1 = df['Habitaciones'] == 1
type(seleccion1)
```

Nos devuelve una serie pandas, que podemos usar como filtro en nuestro dataframe, para que nos devuelva sólo aquellos registros que cumplen la condición

```python
df[seleccion1]
```

```python
seleccion2 = df['Valor'] < 4200
```

```python
df[seleccion2].max()
```

Para aplicar ambas condiciones:

```python
filtro1 = (seleccion1) & (seleccion2)
```

```python
df[filtro1]
```

```python
df1 = df[filtro1]
```

### 2. Apartamentos que tienen al menos 2 dormitorios, un alquiler menor a MXN 10500 y una superficie mayor a 70 m².

```python
filtro2 = (df['Habitaciones'] >= 2) & (df['Valor'] < 10500) & (df['Area'] > 70)
```

```python
df2 = df[filtro2]
```

## Demandas DEV

### Creando Columnas numéricas

El equipo de desarrollo del sitio web de la empresa solicitó la creación de dos nuevas columnas numéricas en nuestra base de datos. En el sitio web de las propiedades, desean presentar información más resumida en relación a los valores. Para esto, estas columnas serán importantes:

```python
datos = pd.read_csv(url, sep=";")
datos.shape
```

#### valor_mensual: Esta columna debe contener los gastos mensuales de cada propiedad, incluyendo el alquiler y el condominio.

```python
datos['valor_mensual'] = datos['Valor'] + datos['Condominio']
datos.head()
```

#### valor_anual: Esta columna debe contener los gastos anuales por propiedad, es decir, el Impuesto más 12 meses de alquiler y condominio.

```python
datos['valor_anual'] = datos['Impuesto'] + datos['valor_mensual'] * 12
datos.tail()
```
