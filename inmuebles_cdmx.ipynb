##### Agrupamos nuestros datos con la herramienta de pandas .groupby('')

* ¿Cuáles son los valores promedio de alquiler por tipo de propiedad?

```python
datos.head()
```

```python
datos[['Tipo','Valor']]
```

```python
type(datos)
```

Si sólo quiero ver algunas columnas

```python
datos.groupby('Tipo')[['Area','Valor']].mean()
```

Si quiero un promedio sólo de las columnas numéricas

```python
datos.groupby('Tipo').mean(numeric_only=True)
```

Si quiero ver cuáles columnas son numéricas

```python
datos.select_dtypes(include='number').columns
```

Como sólo nos interesa la columna 'Tipo' y la columna 'Valor', hacemos lo siguiente, y nos da como salida un dato de tipo series

```python
datos.groupby('Tipo')['Valor'].mean()
```

Para que nos lo de en formato DataFrame, y que también esté ordenado en base a 'Valor', hacemos:

```python
datos.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
```

Para ver mejor los valores, lo graficaremos

```python
df_tipo_precio = datos.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
df_tipo_precio.plot(kind = 'barh', figsize=(12,8), color ='purple')
# kind = 'barh'     para barras horizontales
```

```python
datos.Tipo.unique()
```

```python
# de la lista anterior, extraemos sólo los inmuebles comerciales
inmuebles_comerciales = ['Conjunto Comercial/Sala', 'Edificio Completo', 'Tienda/Salón', 'Casa Comercial', 'Terreno Estándar', 'Cochera/Estacionamiento', 'Galpón/Depósito/Almacén', 'Tienda en Centro Comercial', 'Hotel', 'Loteo/Condominio', 'Industria']
```

Quiero consultar si alguno de los elementos de mi lista está dentro de mi dataset

```python
# uso arroba porque estoy llamando a una variable
datos.query('@inmuebles_comerciales in Tipo')
```

Ahora vamos a consultar la negación, aquellos inmuebles dentro de mi dataset que no son inmuebles comerciales

```python
datos.query('@inmuebles_comerciales not in Tipo')
```

```python
df = datos.query('@inmuebles_comerciales not in Tipo')
df.head()
```

Ahora que tenemos los valores que nos interesan, sí haremos los promedios y los graficaremos

```python
df_tipo_precio = df.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
df_tipo_precio.plot(kind = 'barh', figsize=(12,8), color ='purple')
```

* ¿Cuál es el porcentaje de cada tipo de propiedad en nuestra base de datos?"

```python
df.Tipo.unique()
```

Podemos contar facilmente la cantidad de veces que aparece cada valor dentro de nuestro dataframe, usando:

```python
df.Tipo.value_counts()
```

Así obtenemos los valores prácticamente en porcentajes

```python
df.Tipo.value_counts(normalize=True)
```

```python
porcentaje_tipo = df.Tipo.value_counts(normalize=True).to_frame()
porcentaje_tipo.plot(kind='bar', figsize=(12,8), color = 'green', xlabel='Tipos', ylabel='Porcentajes')
```
